```python
import music21
from music21 import metadata
from music21 import midi
import pathlib

music21.environment.set('musescoreDirectPNGPath', 
                        pathlib.PosixPath('/Applications/MuseScore 3.app/Contents/MacOS/mscore'))


```

```python
# load MIDI file

mf = music21.midi.MidiFile()
mf.open('/Users/jisoolee/Documents/Music/magenta/tlstar.wav.midi')
mf.read()
mf.close()
print(mf)
```

```python
# convert MIDI to Stream

s = midi.translate.midiFileToStream(mf)
print(s)
```

```python
def list_instruments(midi):
    partStream = midi.parts.stream()
    print("List of instruments found on MIDI file:")
    for p in partStream:
        aux = p
        print (p.partName)

list_instruments(s)
```

```python

s.plot('scatter', 'offset', 'pitchClass')
```

```python
timeSignature = s.getTimeSignatures()[0]
music_analysis = s.analyze('key')
print("Music time signature: {0}/{1}".format(timeSignature.beatCount, timeSignature.denominator))
print("Expected music key: {0}".format(music_analysis))
print("Music key confidence: {0}".format(music_analysis.correlationCoefficient))
```
